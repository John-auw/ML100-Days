# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './'
```

```python
import requests

url = 'http://ai100.cupoy.com/file-download/part01/application_train.csv'
r = requests.get(url, allow_redirects=True)
open('application_train.csv', 'wb').write(r.content)

f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python

def isnt_binary(x):
    return len(x.unique()) != 2

numeric = app_train.select_dtypes(include='number') 

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(numeric.columns[numeric.apply(isnt_binary)])
print("Numbers of remain columns: %d\n" % len(numeric_columns))

# 檢視這些欄位的數值範圍
for col in numeric_columns:
    print(col)
#     plt.hist(app_train[col].fillna(0))
#     plt.show()
    app_train[col].plot.box(vert=False)
    plt.show()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
#print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
def ECDF(xs):
    return xs.value_counts().sort_index().cumsum() * 1.0 / len(xs)
#from statsmodels.distributions.empirical_distribution import ECDF
data = np.sort(app_train['AMT_INCOME_TOTAL'])
print(data/data.max())
#exit()
cdf = ECDF(app_train['AMT_INCOME_TOTAL'])

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
