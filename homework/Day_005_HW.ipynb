```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = './data/'
```

```python
import requests

url = 'http://ai100.cupoy.com/file-download/part01/application_train.csv'
r = requests.get(url, allow_redirects=True)
open('application_train.csv', 'wb').write(r.content)

f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv('application_train.csv')
app_train.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

sub_train = pd.DataFrame(app_train['AMT_INCOME_TOTAL'])
mean = np.mean(sub_train)
print('平均數', mean['AMT_INCOME_TOTAL'])
sd = np.std(sub_train, axis=0)
print('標準差', sd['AMT_INCOME_TOTAL'])
sub_train.plot()
```
